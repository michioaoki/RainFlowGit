### 2021-06-15 ～ 2021-09-15のデータ

```python
plot_df = all_df.loc[all_df['date']>='2021-06-15']
plot_df = plot_df.loc[plot_df['date']<='2021-09-15']
plt = plot_df.set_index('date').plot(subplots=True,marker='o',figsize=(28,14),legend=True)
```

### 2020-01-01 ～ 2021-12-31のデータ

```python
plot_df = all_df.loc[all_df['date']>='2020-01-01']
plot_df = plot_df.loc[plot_df['date']<='2021-12-31']
plt = plot_df.set_index('date').plot(subplots=True,marker='o',figsize=(28,14),legend=True)
```

```python
test_df = plot_df
corrs_df = pd.DataFrame(columns=['shift','L1_corr','L2_corr','all_corr'])
for i in range(-5,0,1):
    c1 = np.corrcoef(test_df['rain_mm'][:i],test_df['L1_m3'].shift(i)[:i])[0,1]
    c2 = np.corrcoef(test_df['rain_mm'][:i],test_df['L2_m3'].shift(i)[:i])[0,1]
    cA = np.corrcoef(test_df['rain_mm'][:i],test_df['all'].shift(i)[:i])[0,1]
    shift = i
    #print(shift)
    corrs_df = corrs_df.append({'shift': shift, 'L1_corr':c1, 'L2_corr': c2,'all_corr':cA}, ignore_index=True)
    #print(corrs_df)
    
for i in range(5):
    c1 = np.corrcoef(test_df['rain_mm'][i:],test_df['L1_m3'].shift(i)[i:])[0,1]
    c2 = np.corrcoef(test_df['rain_mm'][i:],test_df['L2_m3'].shift(i)[i:])[0,1]
    cA = np.corrcoef(test_df['rain_mm'][i:],test_df['all'].shift(i)[i:])[0,1]
    shift = i
    corrs_df = corrs_df.append({'shift': shift, 'L1_corr':c1, 'L2_corr': c2,'all_corr':cA}, ignore_index=True)

```

### 降水量と流量のズレを確認　流量を-２シフト、２時間前にすると降水量と相関係数が一番高くなる。すなわち、降水量の影響は２時間後にやってくる。
